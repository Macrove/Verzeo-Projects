```python
import pandas as pd
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('D:/Work/verzeo/StudentsPerformance.csv')
df.head()
```

```python
sns.heatmap(df.corr())
plt.plot()
print('It is easily obervable here that people who have higher reading scores tend to obtain similar results in writing score.')
```

```python
sns.pairplot(df, kind = 'scatter', diag_kind = 'kde', markers = '*')
plt.plot()
print('The above observation can also be noticed here that students have more similarity between reading and writing scores.',
     '\nAlso we can observe that average score in all subject is close to 60.')
```

## Gender

```python
df['gender'].value_counts().plot(kind = 'bar')
plt.show()
```

```python
gender_wise_score = [rows for _, rows in df.groupby('gender')]
gender_wise_score[0].reset_index()
gender_wise_score[1].reset_index()
print('Grouped data according to gender')
```

```python
sns.pairplot(df, hue = 'gender', kind = 'scatter', diag_kind = 'kde', markers = '*')
plt.plot()
print('Inference : Female have lesser standard deviation than Male.',
     ' The average score of male is higher in maths and average score of female is higher than male in reading and writing.')
```

```python
print('Average score of female and male in math :\t'+str(round(gender_wise_score[0]['math score'].mean(),2))+'\t\t'+
      str(round(gender_wise_score[1]['math score'].mean(),2)))
print('Median score of female and male in math :\t'+str(gender_wise_score[0]['math score'].median())+'\t\t'+
      str(gender_wise_score[1]['math score'].median()))
print('Mode score of female and male in math :\t\t'+str(stats.mode(gender_wise_score[0]['math score'])[0][0])+'\t\t'+
      str(stats.mode(gender_wise_score[1]['math score'])[0][0]))
print('std score of female and male in math :\t\t'+str(round(gender_wise_score[0]['math score'].std(),2))+'\t\t'+
      str(round(gender_wise_score[1]['math score'].std(),2)))
print('\n\nINFERENCE : The consistent data above shows that Male have better performance than Female in math.')
```

```python
sns.kdeplot(gender_wise_score[0]['reading score'] ,shade=True , label = 'female')
sns.kdeplot(gender_wise_score[1]['reading score'] ,shade=True , label = 'male')
plt.xlim(0.0,100.0)
plt.title('KDE plot of reading score')
print('Average score of female and male in reading :\t'+str(round(gender_wise_score[0]['reading score'].mean(),2))+'\t\t'+
      str(round(gender_wise_score[1]['reading score'].mean(),2)))
print('Median score of female and male in reading :\t'+str(gender_wise_score[0]['reading score'].median())+'\t\t'+
      str(gender_wise_score[1]['reading score'].median()))
print('Mode score of female and male in reading :\t'+str(stats.mode(gender_wise_score[0]['reading score'])[0][0])+'\t\t'+
      str(stats.mode(gender_wise_score[1]['reading score'])[0][0]))
print('std score of female and male in reading :\t'+str(round(gender_wise_score[0]['reading score'].std(),2))+'\t\t'+
      str(round(gender_wise_score[1]['reading score'].std(),2)))
print('\n\nINFERENCE : The consistent data above shows that Female have better performance than Male in reading.')
```

```python
sns.kdeplot(gender_wise_score[0]['writing score'] ,shade=True , label = 'female')
sns.kdeplot(gender_wise_score[1]['writing score'] ,shade=True , label = 'male')
plt.xlim(0.0,100.0)
plt.title('KDE plot of writing score')
print('Average score of female and male in writing :\t'+str(round(gender_wise_score[0]['writing score'].mean(),2))+'\t\t'+
      str(round(gender_wise_score[1]['writing score'].mean(),2)))
print('Median score of female and male in writing :\t'+str(gender_wise_score[0]['writing score'].median())+'\t\t'+
      str(gender_wise_score[1]['writing score'].median()))
print('Mode score of female and male in writing :\t'+str(stats.mode(gender_wise_score[0]['writing score'])[0][0])+'\t\t'+
      str(stats.mode(gender_wise_score[1]['writing score'])[0][0]))
print('std score of female and male in writing :\t'+str(round(gender_wise_score[0]['writing score'].std(),2))+'\t\t'+
      str(round(gender_wise_score[1]['writing score'].std(),2)))
print('\n\nINFERENCE : The consistent data above shows that Female have better performance than Male in writing.')
```

## Race/Ethnicity

```python
race_wise_score = [rows for _, rows in df.groupby('race/ethnicity')]
for i in range(len(race_wise_score)):
    race_wise_score[i].reset_index()
```

```python
for i in range(len(race_wise_score)):
    sns.kdeplot(race_wise_score[i]['reading score'] ,shade=True , label = chr(ord('A')+i))

plt.xlim(0.0,100.0)
plt.title('KDE plot of reading score')
print('\t\t\t\tREADING SCORE DATA\n')
print('\t Mean\t\t Median\t\tMode\t\tstd')
for i in range(len(race_wise_score)):
    print(chr(ord('A')+i),'\t',round(race_wise_score[i]['reading score'].mean(),2),'\t\t',race_wise_score[i]['reading score'].median(),'\t\t',
     stats.mode(race_wise_score[i]['reading score'])[0][0],'\t\t',round(race_wise_score[i]['reading score'].std(),2))  
print('INFERENCE: The data shows that performance is better in the following order - E > D ~ C > B > A\n'+
'However group A and B are more evenly spread thoughout the score chart than the other groups \nowing to their high standard deviation')
```

```python
for i in range(len(race_wise_score)):
    sns.kdeplot(race_wise_score[i]['writing score'] ,shade=True , label = chr(ord('A')+i))

plt.xlim(0.0,100.0)
plt.title('KDE plot of writing score')
print('\t\t\t\tWRITING SCORE DATA\n')
print('\t Mean\t\t Median\t\tMode\t\tstd')
for i in range(len(race_wise_score)):
    print(chr(ord('A')+i),'\t',round(race_wise_score[i]['writing score'].mean(),2),'\t\t',race_wise_score[i]['writing score'].median(),'\t\t',
     stats.mode(race_wise_score[i]['writing score'])[0][0],'\t\t',round(race_wise_score[i]['writing score'].std(),2))
    
print('INFERENCE: The data shows that performance is better in this following order - E ~ D > C > B > A\n'+
'Also students belongong to group A have relatively poor performance.')
```
